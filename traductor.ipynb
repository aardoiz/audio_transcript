```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
```

```python
tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-es-en")
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-es-en")
```

```python
translator = pipeline("translation_en_to_es", model=model, tokenizer=tokenizer)
```

```python
# Cuidado, si el texto es muy largo el probrama dará un error y tendrás que segmentarlo en trozos más pueños (de máximo 512 caracteres)
es_text = """Esta es una frase de ejemplo: 'me gustan mucho las hamburguesas'
"""

en_text = translator(es_text)[0]['translation_text']
print(en_text)
```
